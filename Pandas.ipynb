## isin()

```python
death_over=df[df['over']>15]
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x=all_batsman>200
```

```python
batsman_list=all_batsman[x].index.to_list()
```

```python
final=df[df['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
```

```python
sr.sort_values(ascending=False).head(5)
```

## merge

```python
matches=pd.read_csv('matches.csv')
deliveries=pd.read_csv('deliveries.csv')
```

```python
matches.shape,deliveries.shape
```

```python
matches.head()
```

```python
deliveries.head()
```

```python
new=deliveries.merge(matches,left_on='match_id',right_on='id')
```

```python
new.shape
```

```python
new.head()
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')
```

## pivot table

```python
six=deliveries[deliveries['batsman_runs'] == 6]
```

```python
pt=six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')
```
